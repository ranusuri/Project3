```python
#dependencies
import pandas as pd
import datetime as dt
from pprint import pprint 
```

```python
#importing csv file and creating dataframes
file = 'Resources/nuforc_reports.csv'
ufo_df = pd.read_csv(file, encoding="ISO-8859-1")
```

```python
ufo_df 
```

```python
ufo_df.dropna(how='any', inplace=True)
ufo_df
```

```python
ufo_df = ufo_df.reset_index(drop=True)
ufo_df
```

```python
clean_ufo_df = ufo_df[["city","state","date_time","shape","city_latitude","city_longitude"]]
clean_ufo_df
```

```python
clean_ufo_df["date_time"] = pd.to_datetime(clean_ufo_df["date_time"]).dt.date
clean_ufo_df.dtypes
```

```python
clean_ufo_df["date_time"] = pd.to_datetime(clean_ufo_df.date_time, format='%Y-%m-%d')
clean_ufo_df.dtypes
```

```python
clean_ufo_df = clean_ufo_df.drop_duplicates()

```

```python
clean_ufo_df = clean_ufo_df[clean_ufo_df['date_time'].dt.year == 2019]
clean_ufo_df = clean_ufo_df.reset_index(drop=True)
clean_ufo_df
```

```python
clean_ufo_df['date_time'] = clean_ufo_df['date_time'].dt.strftime('%m/%d/%Y')
clean_ufo_df.dtypes
```

```python
clean_ufo_df
```

```python
clean_ufo_df = clean_ufo_df.rename(columns={"date_time":"date"})
clean_ufo_df
```

```python
clean_ufo_df.to_excel("UFO_data.xlsx", index=False)
```
